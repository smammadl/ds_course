```python
cap = cv2.VideoCapture(video_path)

while True:
    ret, frame = cap.read()
    if not ret:
        break

    cv2.imshow('Video Playback', frame)

    if cv2.waitKey(25) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

```python
cap = cv2.VideoCapture(video_path)
frame_num = 0

while True:
    ret, frame = cap.read()
    if not ret:
        break
    
    if frame_num % 15 == 0:
        cv2.imshow('video', frame)

        if cv2.waitKey(25) & 0xFF == ord('q'):
            break

    frame_num += 1

cap.release()
cv2.destroyAllWindows()
```
