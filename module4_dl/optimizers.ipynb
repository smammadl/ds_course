# SGD

### ordinary gradient descend

```python
def f(x):
    return x ** 2 + 10

def f_derivative(x):
	return 2*x

def log(x, **kwargs):
	print_str = f"x: {x:.3f} | loss: {f(x):.3f}"
	for k,v in kwargs.items():
		print_str += f" | {k}: {v:.3f}"
	
	print(print_str)

x = 5.0
eta = 0.4

log(x, eta=eta)

for i in range(10):
	x = x - eta * f_derivative(x)
	log(x, eta=eta)
```

### gd with momentum

```python
x = 5.0
eta = 0.4
gamma = 0.1
v = 0

log(x, eta=eta)

for i in range(10):
	v = eta * f_derivative(x) + gamma * v
	x = x - v
	log(x, eta=eta)
```

### nesterov momentum

```python
x = 5.0
eta = 0.4
gamma = 0.1
v = 0

log(x, eta=eta)

for i in range(10):
	x_ahead = x - gamma * v
	v = eta * f_derivative(x_ahead) + gamma * v
	x = x - v
	log(x, eta=eta)
```

# Adagrad

$$\theta_{t+1,i} = \theta_{t,i} - \frac{\eta}{\sqrt{G_{t,ii} + \epsilon}} \cdot g_{t,i}$$
$$G_{t,ii} = \sum_{\tau=1}^{t} g_{\tau,i}^2$$

```python
x = 5.0
eta_base = 0.5
epsilon = 1e-8
g_total = 0

log(x, grad=f_derivative(x), eta=float('inf'))

for i in range(10):
	g_total += f_derivative(x)**2
	eta_adaptive = eta_base / (g_total + epsilon)**(1/2)
	x = x - eta_adaptive * f_derivative(x)
	log(x, grad=f_derivative(x), eta=eta_adaptive)
```

# RMSProp

$$E[g^2]_t = \rho \cdot E[g^2]_{t-1} + (1 - \rho) \cdot g_t^2$$
$$\theta_{t+1} = \theta_t - \frac{\eta}{\sqrt{E[g^2]_t + \epsilon}} \cdot g_t$$

```python
x = 5.0
eta_base = 0.7
epsilon = 1e-8
second_moment = 0
rho = 0.9

log(x, grad=f_derivative(x), second_moment=second_moment, eta=float('inf'))

for i in range(20):
	second_moment = rho * second_moment + (1 - rho) * f_derivative(x)**2
	eta_adaptive = eta_base / (second_moment + epsilon)**(1/2)
	x = x - eta_adaptive * f_derivative(x)
	log(x, grad=f_derivative(x), second_moment=second_moment, eta=eta_adaptive)
```

# Adadelta

$$\Delta \theta_t = -\frac{\sqrt{E[\Delta \theta^2]_{t-1} + \epsilon}}{\sqrt{E[g^2]_t + \epsilon}} \cdot g_t$$
$$\theta_{t+1} = \theta_t + \Delta \theta_t$$

```python
x = 5.0
epsilon = 1e-6
second_moment_grad = 0
delta_x = 0
second_moment_delta_x = 0
rho = 0.95

log(x, grad=f_derivative(x), second_moment_grad=second_moment_grad, second_moment_delta_x=second_moment_delta_x, delta_x=delta_x)

for i in range(20):
	second_moment_grad = rho * second_moment_grad + (1 - rho) * f_derivative(x)**2
	second_moment_delta_x = rho * second_moment_delta_x + (1 - rho) * delta_x**2
	delta_x = - ((second_moment_delta_x + epsilon) / (second_moment_grad + epsilon))**(1/2) * f_derivative(x)
	x = x + delta_x
	log(x, grad=f_derivative(x), second_moment_grad=second_moment_grad, second_moment_delta_x=second_moment_delta_x, delta_x=delta_x)
```

# Adam

$$m_t = \beta_1 \cdot m_{t-1} + (1 - \beta_1) \cdot g_t$$

$$v_t = \beta_2 \cdot v_{t-1} + (1 - \beta_2) \cdot g_t^2$$

$$\hat{m}_t = \frac{m_t}{1 - \beta_1^t}$$

$$\hat{v}_t = \frac{v_t}{1 - \beta_2^t}$$

$$\theta_{t+1} = \theta_t - \frac{\eta}{\sqrt{\hat{v}_t} + \epsilon} \cdot \hat{m}_t$$


```python
x = 5.0
eta_base = 0.4
epsilon = 1e-6
first_moment = 0
beta1 = 0.9
second_moment = 0
beta2 = 0.999

log(x, grad=f_derivative(x), m=first_moment, m_hat=0, v=second_moment, v_hat=0, update=eta_adaptive)

for i in range(30):
	first_moment = beta1 * first_moment + (1 - beta1) * f_derivative(x)
	first_moment_corrected = first_moment / (1 - beta1**(i+1))

	second_moment = beta2 * second_moment + (1 - beta2) * f_derivative(x)**2
	second_moment_corrected = second_moment / (1 - beta2**(i+1))

	eta_adaptive = eta_base / (second_moment_corrected**(1/2) + epsilon) * first_moment_corrected

	x = x - eta_adaptive
	log(x, grad=f_derivative(x), m=first_moment, m_hat=first_moment_corrected, v=second_moment, v_hat=second_moment_corrected, update=eta_adaptive)
```
