
$$\nabla J(\mathbf{\theta}) = \frac{1}{m} \mathbf{X}^T (\mathbf{H} - \mathbf{Y})$$

```python
from sklearn.preprocessing import add_dummy_feature
eta = 0.1
n_epoch = 60000
X_train_b = add_dummy_feature(X_train)
m, n = X_train_b.shape
m, K = y_train.shape

def softmax(Z):
    return np.exp(Z) / np.sum(np.exp(Z), axis=1, keepdims=True)

rng = np.random.default_rng(seed=123)
Theta = rng.standard_normal(size=(n, K))

for epoch in range(n_epoch):
	Z = X_train_b @ Theta
	H = softmax(Z)
	gradient = 1/m * X_train_b.T @ (H - y_train)
	Theta = Theta - eta * gradient
Theta
```

```python
X_test_b = add_dummy_feature(X_test)
Z = X_test_b @ Theta
H = softmax(Z)
y_pred = np.argmax(H, axis=1)
y_pred
```

```python
y_act = np.argmax(y_test, axis=1)
y_act
```

```python
acc = (y_pred == y_act).sum() / len(y_pred)
acc
```

```python
from sklearn.linear_model import LogisticRegression
clf = LogisticRegression()
clf.fit(X_train, np.argmax(y_train, axis=1))
y_pred = clf.predict(X_test)
y_pred
```
