```python
bosh_list = []

for s in range(-5,5):
	if s%2==0:
		if s>0:
			kv = s**2
			bosh_list.append(kv)
		else:
			bosh_list.append(s)
    
bosh_list
```

# error handling

```python
x = 5
y = '2'

try:
    print(x/y)
except TypeError:
    print('x-in ve ya y-in tipi dogru deyil')
    if type(x) != int:
        x = int(x)
    elif type(y) != int:
        y = int(y)
    print(x/y)
except Exception as e:
    print(f'nese problem oldu: {e}')
```

```python
x = 5
y = 0

try:
    print(x/y)
except TypeError:
    print('x-in ve ya y-in tipi dogru deyil')
except Exception as e:
    print(f'nese problem oldu: {e}')
else: 
    print('error cixmadi')
finally:
    print('proqrami icra etdik qutardiq')
```
