```python
class Heyvan:
	yasayir = True
	gorur = True
	hereket_edir = False

	def hereket_et(self):
		self.hereket_edir = True
		return 'hereket edirem'

class SuHeyvani(Heyvan):
    yasayish_yeri = 'su'
    
class QuruHeyvani(Heyvan):
    yasayish_yeri = 'quru'
    
class Baliq(SuHeyvani):
	qelsemesi_var = True

	def __init__(self, b_adi, b_rengi='yashil', b_balacadir=True):
		self.ad = b_adi
		self.reng = b_rengi
		self.balacadir = b_balacadir

	def uz(self):
		return 'uzurem'

	def hereket_et(self):
		normalda_neynemeliydi = super().hereket_et()
		if self.ad == 'felaket':
			return normalda_neynemeliydi + ' akvarium dagildi'
		else:
			return normalda_neynemeliydi + ' plus ' + self.uz()

b1 = Baliq('zumrud')
b2 = Baliq('felaket', 'boz', False)
b3 = Baliq('şemi', 'sarı')
b2.hereket_et()
```

```python
class Mashin:
	tekeri_var = True
	rolu_var = True
	_yanacaq = 'benzin'
	__suret = 0

	def __init__(self, m_markasi, m_modeli):
		self.marka = m_markasi
		self.model = m_modeli

	def get(self):
		return 'gedirem'
	
	def suretlen(self):
		self.__suret = self.__suret + 10
		return f'ok suretlendik, yeni suret: {self.__suret}'
	
	def sureti_goster(self):
		return self.__suret

m = Mashin('toyota', 'prius')
print(m.sureti_goster())
print(m.suretlen())
print(m.suretlen())
print(m.sureti_goster())
```
