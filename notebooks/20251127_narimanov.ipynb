```python
class DersSinifi:
	telebeler_var = True

	def __init__(self, say):
		self.say = say

	@property
	def telebeler(self):
		return DersSinifi.telebeler_var

s = DersSinifi(9)
s.telebeler
```

```python
# import dataclasses
# @dataclasses.dataclass

from dataclasses import dataclass
@dataclass
class DersSinifi:
    say : int
    telebeler_var : bool = True
    
s = DersSinifi(9)
s.say
```
