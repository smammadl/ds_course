```python
def salamla(ad):
    return f"salam {ad}"

def salamla(ad1, ad2):
    return f"salam {ad1},{ad2}"

def salamla(*adlar):
    return 'salam ' + ', '.join(adlar)

salamla('ali', 'hesen', 'huseyn', 'lale')
```

```python
def melumat_ver(ad='ali', yash=25):
    print(f'ad : {ad}')
    print(f'yash : {yash}')
    
def melumat_ver(**atributlar):
    for k,v in atributlar.items():
        print(f"{k}: {v}")
        
def melumat_ver(*args, **kwargs):
    for nese in args:
        print(nese)
    for k,v in kwargs.items():
        print(f"{k}: {v}")

melumat_ver('ali', 'huseynov', 'hesen', yash=25, boy=180)
```

```python
class Riyaziyyat:
	pi = 3.14

	def __init__(self, r):
		self.radius = r

	@staticmethod
	def vur(a, b):
		return a*b

	@staticmethod
	def kok_al(sayi):
		return sayi**(1/2)
	
	@classmethod
	def sahe_tap(cls,radius):
		return cls.pi * radius**2
	
	@property
	def diameter(self):
		return 2*self.radius
    
# Riyaziyyat.kok_al(9)
# Riyaziyyat.sahe_tap(5)
r = Riyaziyyat(r=3)
r.diameter
```
