```python
sheherler = {
	'Baku' : '2,344,900',
	'Sumgait' : '427,000',
	'Ganja' : '330,700',
	'Xırdalan' : '195,700',
	'Mingachevir' : '102,900'	
}

sheherler = {sh:int(e.replace(',', '')) for sh,e in sheherler.items()}
sheher = sheherler.keys()
ehali = sheherler.values()

df = pd.DataFrame({
	'ehali': ehali, 
	'boyukluk' : ['boyuk', 'boyuk', 'boyuk', 'kicik', 'kicik']
},
	index = sheher,
)

df
```

```python
yeni_ehali = df['ehali'].apply(lambda r: r*1.2)
df.loc[:, 'ehali'] = yeni_ehali
df
```

```python
df['boyukluk'].mask(df['ehali'] < 5e5, 'orta')
```
