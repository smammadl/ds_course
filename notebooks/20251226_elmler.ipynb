```python
df.groupby(['Continent'], as_index=False)['2022 Population'].sum()
df.groupby(['Continent'])['2022 Population'].sum()
df.groupby(['Continent'])['Area (km²)'].mean()
```

```python
df['2030 Population'] = df.loc[:, '2022 Population'] * df.loc[:, 'Growth Rate']**8

def kategoriyalashdir(pop : float) -> str:
	if pop > 100e6:
		return 'large'
	elif pop > 25e5:
		return 'mid'
	else:
		return 'small'
    
df['2030_cat'] = df['2030 Population'].apply(kategoriyalashdir)
df.head(2)
```

```python
df.groupby(['Continent', '2030_cat'])['Country'].count().to_frame().reset_index()
```

```python
df.groupby('Continent')[
    ['2022 Population', '2030 Population']
    ].sum()

df.groupby('Continent').agg({
    '2022 Population' : 'sum', 
    '2030 Population' : 'sum', 
})

df.groupby('Continent').agg(
    pop_sum_2022 = ('2022 Population', 'sum'),
    pop_sum_2030 = ('2030 Population', 'sum'),
)
```

```python
# df.groupby('Continent')['2022 Population'].sum()
# df.groupby('Continent')['2022 Population'].transform('sum')
df['cont_pop'] = df.groupby('Continent')['2022 Population'].transform('sum')
df.head(4)
```

```python
def boyuk_qiteleri_saxla(balaca_df):
    return balaca_df['2022 Population'].sum() > 1e9

df.groupby('Continent').filter(boyuk_qiteleri_saxla)
```
