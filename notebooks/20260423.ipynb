```python
class TehsilMuessisesi:
	def __init__(self, ad, unvan):
		self.ad = ad
		self.unvan = unvan

class Kurs(TehsilMuessisesi):
	def __init__(self, ad, unvan, proqramlar):
		super().__init__(ad, unvan)
		self.proqramlar = proqramlar

k1 = Kurs(
	ad='div', 
	unvan='elmler', 
	proqramlar=['DS', 'Frontend', 'Bakend', 'Fullstak']
	)
```

```python
from dataclasses import dataclass

@dataclass
class TehsilMuessisesi:
    ad:str
    unvan:str
    
@dataclass
class Kurs(TehsilMuessisesi):
    proqramlar:list[str]
    

k1 = Kurs(
	ad='div', 
	unvan='elmler', 
	proqramlar=['DS', 'Frontend', 'Bakend', 'Fullstak']
	)
k1.unvan
```
