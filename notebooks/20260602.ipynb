```python
import pandas as pd
```

```python
from sqlalchemy import create_engine

query_emp = """
SELECT * 
FROM employees   
limit 100;
"""
query_sal = """
SELECT * 
FROM salaries  
limit 100;
"""

con = create_engine("postgresql+psycopg2://postgres:postgres@localhost/employees")

df_emp = pd.read_sql_query(query_emp, con)
df_sal = pd.read_sql_query(query_sal, con)
```
