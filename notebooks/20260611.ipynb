```python
def f(x,y):
	return np.sin(x) ** 10 + np.cos(10+y*x) + np.cos(x)
```

```python
x = np.linspace(0,5,100)
y = np.linspace(0,5,100)

X,Y = np.meshgrid(x,y)
X.shape, Y.shape

Z = f(X,Y)
```

```python
plt.contour(X,Y,Z, colors='black')
```

```python
plt.contour(X,Y,Z,10, cmap='jet')
```

```python
contours = plt.contour(X,Y,Z,colors='black')
plt.clabel(contours, inline=True, fontsize=7)
plt.imshow(Z, extent=[0,5,0,5], origin='lower',cmap='jet',alpha=0.75)
plt.colorbar()
```
