```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path
```

```python
x = np.linspace(0,10,100)

# plt.plot(x, np.sin(x), linestyle='--', color='red', marker='o')
plt.scatter(x, np.sin(x), color='red', marker='o')
plt.title('sin graph')
plt.xlabel('x')
plt.ylabel('sin')
```

```python
x = np.linspace(0,6,50)
y = np.linspace(0,5,60)

X,Y = np.meshgrid(x,y)

def f(x, y):
    return np.sin(x) ** 10 + np.cos(10 + y * x) * np.cos(x)

Z = f(X,Y)

countour = plt.contour(X, Y, Z, levels=3, colors='black')
plt.clabel(countour, fontsize=7)
plt.imshow(Z, origin='lower', extent=[0,6,0,5], alpha=0.5, cmap='jet')
plt.colorbar()
plt.show()
```
