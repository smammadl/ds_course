```python
class Heyvan:
	def __init__(self, h_adi):
		self.ad = h_adi

	def ozunu_tanit(self):
		return f"menim adim {self.ad}."

class It(Heyvan):
	def __init__(self, i_adi):
		super().__init__(i_adi)

	def ozunu_tanit(self):
		normalda_qaytarmaliydi = super().ozunu_tanit()
		eslinde_qaytar = normalda_qaytarmaliydi + ' Men item.'
		return eslinde_qaytar
		
a = It('feraxet')
a.ozunu_tanit()
```

```python
class Heyvan:
	def __init__(self, h_adi, h_novu):
		self.ad = h_adi
		self.nov = h_novu

	def ozunu_tanit(self):
		return f"menim adim {self.ad}. Menim novum: {self.nov}"

class It(Heyvan):
	def __init__(self, i_adi):
		nov = It.__name__
		super().__init__(i_adi, nov)
		
a = It('feraxet')
a.ozunu_tanit()
```

```python
class Mashin:
	muherriki_var = True

	def __init__(self, m_markasi, m_modeli=None, m_ili=None, m_rengi=None):
		self.isleyir = False
		self.model = m_modeli
		self.marka = m_markasi
		self.reng = m_rengi
		self.il = m_ili

	def ishe_sal(self):
		self.isleyir = True

	def __str__(self):
		return f"Mashin(Marka: {self.marka}, Model: {self.model})"	

	def __repr__(self):
		return f"Mashin(Marka: {self.marka}, Model: {self.model})"	
	
	def __eq__(self, other):
		return self.marka == other.marka
	
	def __del__(self):
		print(f'{self.marka} mashin silindi')


a = Mashin('tefal')
del a
```
