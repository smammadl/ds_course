```python
fig,axs = plt.subplots(3,3)

for i, qutu in enumerate(axs.flatten()):
	X_reshaped = X[[i+100], :].reshape(28,28)
	qutu.imshow(X_reshaped, cmap='gray')
	qutu.set_title(y[i+100])
plt.show()
```

```python
pd.set_option("display.max_columns", None)

pd.DataFrame(X)
```

```python
X_scaled = X/255
X_scaled
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.5, random_state=42)
X_train.shape, y_train.shape
```

```python
from sklearn.linear_model import LogisticRegression
log_reg = LogisticRegression()
log_reg.fit(X_train, y_train)
y_pred = log_reg.predict(X_test)
```

```python
y_pred
```

```python
y_test
```

```python
(y_test == y_pred).mean()
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
cm
```

```python
cm.sum(axis=1)
```

```python
from sklearn.metrics import ConfusionMatrixDisplay

fig, axs = plt.subplots(figsize=(6,6))
disp = ConfusionMatrixDisplay(cm, display_labels=mnist['categories']['class'])
disp.plot(ax=axs, cmap='Blues')
plt.show()
```
