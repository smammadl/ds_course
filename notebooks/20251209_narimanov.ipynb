```python
def f(pop):
	if pop > 100e6:
		return 'boyuk olke'
	else:
		return 'kicik olke'

s_pop_yeni = s_pop.apply(f)
df['2030_population_index'] = s_pop_yeni
df.head(3)
```

```python
s_cont = df['Continent']

def inkisaf_sev_gore_ayir(qite):
    if qite == 'Europe' or qite == 'North America':
        return 'inkisaf etmis'
    else:
        return 'inkisaf eden'

s_cont_yeni = s_cont.apply(inkisaf_sev_gore_ayir)
df['ink_seviyesi'] = s_cont_yeni
df
```
