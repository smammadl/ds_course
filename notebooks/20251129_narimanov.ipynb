```python
x = ['a', 'b', 'c']
y = x.copy()
print(x is y)
y[0] = 'Z'
print(y)
print(x)
```

```python
import copy
x = [['a','b','c'], [1,2,3]]
y = copy.deepcopy(x)
print(x is y)
x[0][0] = 'Z'
print(x)
print(y)
```

```python
x = [1, 2, 3]
y = x
y = ['a', 'b', 'c'] # This will create a new object in memory and assign it to y
print(x) # Output: [1, 2, 3]
print(y) # Output: ['a', 'b', 'c']
print(x is y) # Output: False
```
