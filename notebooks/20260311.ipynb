```python
from sklearn.datasets import load_iris
iris = load_iris()
X = iris['data']
y = iris['target']
y = (y != 0).astype('int')[:, np.newaxis]
```

```python
from sklearn.preprocessing import add_dummy_feature
X_b = add_dummy_feature(X)
X_b.shape
```


$$\nabla J(\mathbf{\theta}) = \frac{1}{m} \mathbf{X}^T (\mathbf{h} - \mathbf{y})$$

$$\theta_{yeni} = \theta_{kohne} - \eta \ \nabla J(\mathbf{\theta})$$

```python
from sklearn.model_selection import train_test_split

X_b, X_test_b, y, y_test = train_test_split(
    X_b, y, test_size=0.2
    )
```

```python
m,n = X_b.shape
eta = 0.1
n_epochs = 50000

rng = np.random.default_rng(seed=123)
theta = rng.standard_normal(size=(n,1))

for epoch in range(n_epochs):
	z = X_b@theta
	h = sigmoid(z)
	gradient = 1 / m * X_b.T @ (h - y)
	theta = theta - eta * gradient

theta
```

```python
z = X_test_b @ theta
h = sigmoid(z)
y_pred = np.where(h>0.5, 1, 0)
(y_test == y_pred).flatten().sum()
```
