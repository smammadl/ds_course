```python
rng = np.random.default_rng(0)
data = rng.multivariate_normal(
    mean=[0,0],
    cov=[[5,3],
         [3,2]],
	size=1000
)
df = pd.DataFrame(data, columns=['x', 'y'])

sns.set_theme()
plt.hist(df['x'], alpha=0.5, density=True, label='x')
plt.hist(df['y'], alpha=0.5, density=True, label='y')
plt.legend();
```

```python
sns.kdeplot(data=df, fill=True)
```

```python
sns.kdeplot(data=df, x='x', y='y')
```

```python

sns.histplot(data=df, kde=True)
```

```python
tips = sns.load_dataset('tips')
tips['tip_pct'] = tips['tip'] / tips['total_bill'] * 100
tips.head(2)
```

```python
grid = sns.FacetGrid(
    data=tips,
    row='sex',
    col='time',
    margin_titles=True
    )
b = np.linspace(0,50,20)
grid.map(plt.hist, 'tip_pct', bins=b)
```

```python
cp = sns.catplot(data=tips, x='day', y='total_bill', hue='sex', kind='box')
cp.set_axis_labels(x_var='Day', y_var='Total Bill')
```
