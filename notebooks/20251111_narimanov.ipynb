```python
fruits = ['alma', 'armud', 'qarpiz']

if 'chiyelek' in fruits:
    print('chiyelike fruitsdadir')
elif 'mandarin' in fruits:
    print('mandarin fruitsdadir')
else:
    print('meyve yoxdur')
```

```python
age = 20

if age >= 20: # True ctrl shift [
	print('yetkin')
	if age > 60:
		print('yashli')
	elif age > 40:
		print('orta yashli')
	else:
		print('genc')
elif age <20:
	print('askjdfla')
else:
    print('yeniyetme')
```

```python
# match
day = 'Holiday'

match day:
    case 'Monday':
        print('monday')
    case 'Tuesday':
        print('tuesday')
    case 'Wednesday':
        print('x')
    case 'Thursday':
        print('y')
    case 'Friday':
        print('z')
    case 'Saturday':
        print('w')
    case 'Sunday':
        print('w')
    case _:
        print('invalid weekday')
```

```python
name_age = [('ali', 20), ('hesen', 21), ('xalide', 30), ('gulnar', 25)]

for i, (n,a) in enumerate(name_age):
    if i == 1:
	    break
    print(i)
    print(n,a)
    print('---')
```

```python
name_age = [('ali', 20), ('hesen', 21), ('xalide', 30), ('gulnar', 25), ('hesen', 51)]

for i, (n,a) in enumerate(name_age):
    if n == 'hesen' and a==21:
        continue
    print(i)
    print(n,a)
    print('---')
```
