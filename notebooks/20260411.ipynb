```python
from pathlib import Path
import shutil

def kocur(from_dir, to_dir):
	files = list(from_dir.iterdir())

	for f in files:
		new_path = to_dir / f.name
		shutil.copy2(f, new_path)
		f.unlink()

	return list(to_dir.iterdir())

from_dir = Path(r"C:\Users\user\Desktop\salam")
to_dir = Path(r"C:\Users\user\Desktop\necesen")

kocur(from_dir, to_dir)
```
