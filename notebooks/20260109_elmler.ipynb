```python
rng = np.random.default_rng(0)
# data1 = rng.normal(0,5,100)
# data2 = rng.normal(0,2,100)
# df = pd.DataFrame(data={
#     'data1' : data1,
#     'data2' : data2
# })
data = rng.multivariate_normal(
    mean=[0,0],
    cov=[[5,3],
         [3,2]],
    size=1000
)
df = pd.DataFrame(data, columns=['a', 'b'])
df.head(2)
```

```python
plt.hist(
    df['a'], 
    bins=20, 
    alpha=0.5,
    edgecolor='black'
    )
plt.hist(
    df['b'],
    bins=20, 
    alpha=0.5,
    edgecolor='black'
    )
```

```python
sns.histplot(data=df, bins=20, kde=True)
```

```python
sns.histplot(data=df, x='a', y='b')
```

```python
sns.scatterplot(data=df, x='a', y='b')
```

```python
sns.kdeplot(data=df, x='a', y='b')
plt.xlim([-10,10])
plt.ylim([-10,10])
```

```python
tips = sns.load_dataset('tips')
tips['tip_pct'] = (tips['tip'] / tips['total_bill'] * 100).round(2)
tips.head(2)
```

```python
# plt.style.available
plt.style.use('seaborn-v0_8-whitegrid')
```

```python
sns.set_theme()
```

```python
grid = sns.FacetGrid(tips, row='sex', col='time', margin_titles=True)
grid.map(plt.hist, 'tip_pct', bins=np.linspace(0, 50, 20))
```
