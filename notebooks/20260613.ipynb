```python
x = np.linspace(0,10,50)

fig, axs = plt.subplots(nrows=2, ncols=2, sharex=True)
axs[0,1].plot(x,np.sin(x), color='blue')
axs[1,0].plot(x,np.cos(x), color='red')
axs[0,0].plot(x,np.tan(x), color='red')
axs[1,1].remove()
```

```python
axs.flat
```

```python
x = np.linspace(0,10,50)

fig, axs = plt.subplots(nrows=2, ncols=2, sharex=True)

for i in range(2):
	for j in range(2):
		axs[i,j].plot(x, np.sin(x+i+j))
```

```python
x = np.linspace(0,10,50)

fig, axs = plt.subplots(nrows=2, ncols=2, sharex=True)

for i, ax in enumerate(axs.flat):
	ax.plot(x, np.sin(x+i))
```
