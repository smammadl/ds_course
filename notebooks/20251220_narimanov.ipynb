```python
import pandas as pd
import numpy as np
import pathlib
# from pathlib import Path
```

```python
csv_path = pathlib.Path(r"C:\Users\user\Desktop\div\data\1.04. Real-life example.csv")
df = pd.read_csv(csv_path)
df
```

```python
df_yeni = df.copy()
df_yeni.set_index('Brand', inplace=True)

yeni_cols = [
['General', 'General', 'General', 'General', 'Main', 'Main', 'Main', 'Main'],
['Price', 'Body', 'Mileage', 'EngineV', 'Engine Type', 'Registration', 'Year', 'Model']
]

df_yeni.columns = yeni_cols
df_yeni.head(3)
```

```python
mask = df_yeni.loc[:, ('General', 'Body')] != 'sedan'
df_yeni.loc[mask]
```

```python
df.head(3)
```

```python
mask_c = df['Body'] == 'crossover'
mask_h = df['Body'] == 'hatch'
# df.loc[mask_c | mask_h, :].groupby(['Brand', 'Body'])['Price'].sum()
df.loc[mask_c | mask_h, :].groupby(['Brand', 'Body']).agg({
    'Price' : 'sum',
    'Mileage' : 'mean',
    'Model' : 'nunique',
    # 'Model' : 'unique',
})
```

```python
df.groupby(['Brand', 'Body'])['Price'].sum()
```

```python
s_brand_avg = df.groupby('Brand')['Price'].transform('mean')
df['brand_avg'] = s_brand_avg
df
```

```python
df.groupby('Brand')['Price'].mean()
df2 = df.groupby('Brand', as_index=False).agg({
    'Price' : 'mean'
})
df3 = df.drop('brand_avg', axis='columns')
pd.merge(df3, df2, how='left', on='Brand').head(3)
```
