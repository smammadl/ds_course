```python
def salam_de(ad:str):
	if ad=='ali':
		qaytarilan_cumle = f"salam {ad}"
	elif ad=='hesen':
		qaytarilan_cumle = f"salam e"
	else:
		qaytarilan_cumle = f"hervaxtiniz xeyir"

	return qaytarilan_cumle

salam_de('lale')
```

```python
def salam_de(adamlar):
    salam_cumlesi = 'salam '
    for adam in adamlar:
        salam_cumlesi = salam_cumlesi + f"{adam}, "
    return salam_cumlesi

salam_de(['ali', 'hasan', 'lale'])
```

```python
def len_ozumun(l):
	count = 0
	for elem in l:
		count += 1
	return count

len_ozumun([1,34,112,1,23,1])
```
