```python
df = pd.DataFrame({
	'Team': ['A', 'A', 'B', 'B', 'C', 'C'],
	'Points': [10, 15, 10, 20, 10, 5],
	'Assists': [5, 7, 5, 10, 5, 2]
})
df
```

```python
def f(balaca_df):
    assists_sum = balaca_df['Assists'].sum()
    if assists_sum > 10:
        balaca_df['Points'] = balaca_df['Points'] * 2
    return balaca_df

df.groupby('Team', group_keys=False).apply(f)
```

```python
df_balaca = df.groupby('Team').get_group('A')
assists_sum = df_balaca['Assists'].sum()
if assists_sum > 10:
    df_balaca['Points'] = df_balaca['Points'] * 2
df_balaca
```
