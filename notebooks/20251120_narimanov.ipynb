```python
import math

class Cevre:
	pi = math.pi

	def __init__(self, yeni_radius):
		self.radius = yeni_radius
		self.sahe = Cevre.pi * (self.radius**2)
		self.uzunluq = 2 * Cevre.pi * self.radius
	
	def radiusuDeyish(self,yeni_radius):
		self.radius = yeni_radius
		self.sahe = Cevre.pi * (self.radius**2)
		self.uzunluq = 2 * Cevre.pi * self.radius

c1 = Cevre(1)
print(c1.radius)
print(c1.sahe)
print(c1.uzunluq)
c1.radiusuDeyish(4)
print(c1.radius)
print(c1.sahe)
print(c1.uzunluq)
```
