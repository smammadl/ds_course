```python
import os

class PathClone:
	def __init__(self, object_path):
		self.path = object_path
		self.clean_path = object_path.replace('\\', '/')

	def __repr__(self):
		return f"WindowsPath(\'{self.clean_path}\')"
	
	@property
	def name(self):
		return self.clean_path.split('/')[-1]

	@staticmethod
	def cwd():
		return os.getcwd()
		

img_path = PathClone(r'C:\Users\user\Desktop\div\test\shekil.png')
# img_path.name
PathClone.cwd
# img_path.cwd()
```
