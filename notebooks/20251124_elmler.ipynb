```python
def cutdur(eded):
    if eded%2==0:
        return True
    else:
        return False

def hamisi_cutdur(l):
	for eded in l:
		if cutdur(eded):
			continue
		else:
			return False
	return True

hamisi_cutdur([2,4,5,6])
```

```python
def cutdur(eded):
    if eded%2==0:
        return True
    else:
        return False

def cut_tek(l):
    bos_list = []
    for eded in l:
        if cutdur(eded):
            bos_list.append('cut')
        else:
            bos_list.append('tek')
    return bos_list

cut_tek([1,2,3,4])
```
