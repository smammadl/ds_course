```python
from sklearn.datasets import load_iris
data = load_iris()  
data

X = data['data']
y = data['target']
# y = data['target_names']
# pd.Series(y)
```

```python
from sklearn.preprocessing import OneHotEncoder
one_hot = OneHotEncoder(sparse_output=False)
y_onehot = one_hot.fit_transform(y[:, np.newaxis,])
y_onehot
```

```python
df= pd.DataFrame(np.concat([X, y_onehot], axis=1), columns=data['feature_names']+['class0', 'class1', 'class2'])
df
```

```python
from sklearn.preprocessing import add_dummy_feature
X_b = add_dummy_feature(X)
X_b
```

```python
from scipy.special import softmax
n_epoch=200
theta = np.random.standard_normal((5,3))
eta = 0.1
m = len(X_b)

for epoch in range(n_epoch):
	Z = X_b @ theta # (150, 5) x (5, 3) = (150,3)
	H = softmax(Z, axis=1)
	gradient = (1/m) * X_b.T @ (H - y_onehot)
	theta = theta - eta * gradient

Z = X_b @ theta # (150, 5) x (5, 3) = (150,3)
H = softmax(Z, axis=1)
y_pred = H.argmax(axis=1)
y_pred
```

```python
(y == y_pred).mean()
```

```python
log_reg = LogisticRegression(max_iter = 200)
log_reg.fit(X, y)
y_pred = log_reg.predict(X)
(y == y_pred).mean()
```

```python
acc = accuracy_score(y, y_pred)
acc
```
