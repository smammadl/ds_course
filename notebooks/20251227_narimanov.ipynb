```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
x = np.linspace(0,10,50)

plt.figure(figsize=(5,4))
plt.plot(x, np.sin(x), color='b', linestyle='-', marker='o', label='sin')
plt.plot(x, np.cos(x), color='k', linestyle='--', marker='.', label='cos')

plt.title('Sin Cos Graph')

plt.xlabel('x')
plt.ylabel('trig(x)')

plt.legend(loc='lower right')
plt.savefig('trig.png')
```

```python
import yfinance as yf
import datetime

data = yf.download('NVDA', '2022-01-01', datetime.datetime.today())
df = data.droplevel(1, axis='columns')
df.reset_index(inplace=True)
df.head(3)
```

```python
df_sample = df.sample(n=100).sort_values('Date')
df_sample
```

```python
plt.plot(df['Date'], df['Close'])
plt.gcf().autofmt_xdate()

plt.plot(df_sample['Date'], df_sample['Close'], linestyle='', marker='.')
```

```python
plt.plot(df['Date'], df['Close'])
plt.gcf().autofmt_xdate()

df['Close_std30'] = df['Close'].rolling(window=90).std()
plt.fill_between(
    df['Date'], 
    df['Close'] - df['Close_std30'], 
    df['Close'] + df['Close_std30'],
    color='gray',
    alpha=0.5
)
```
