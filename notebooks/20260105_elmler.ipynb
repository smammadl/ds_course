```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from pathlib import Path
```

```python
x = np.linspace(0,10,100)

plt.plot(x, np.sin(x), linestyle='-', marker='o', color='red', label='sin(x)')
plt.plot(x, np.cos(x), linestyle='--', marker='', color="#351C5D", label='cos(x)') 
plt.title('Trig Graphs')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
# linestyle - -- -. :
# marker 0 . , > < ^ v * 
# color 'red' 'r' (1, 0, 0) #FF0000
```

```python
x = np.linspace(0,10,100)

plt.scatter(x, np.sin(x),  marker='o', color='red', label='sin(x)')
plt.scatter(x, np.cos(x), marker='.', color="#351C5D", label='cos(x)') 
plt.title('Trig Graphs')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
# marker 0 . , > < ^ v * 
# color 'red' 'r' (1, 0, 0) #FF0000
```

```python
x = np.linspace(0,5,60)
y = np.linspace(-2,2,50)
X, Y = np.meshgrid(x,y)

def f(x, y):
    return np.sin(x) ** 10 + np.cos(10 + y * x) * np.cos(x)

Z = f(X,Y)

contour = plt.contour(X, Y, Z, levels=6, colors='k')
plt.clabel(contour, fontsize=9)
plt.imshow(Z, extent=[0,5,-2,2], origin='lower', 
           cmap='magma', alpha=0.8)
```
