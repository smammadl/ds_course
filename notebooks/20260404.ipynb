```python
camp = [
    'xosh gelmisen', 
    'xosh getmisen',
    'konullusen',
    'konulsuzsen',
    '1 al besh ode'
    ]

disc = {
    'xosh gelmisen' : {
        'aktivdir' : True, 
        'endirim' : 0.3
        },
    'xosh getmisen' : {
        'aktivdir' : False, 
        'endirim' : 0.2
        },
    'konulsuzsen' : {
        'aktivdir' : True, 
        'endirim' : 0.15
        },
    'diger' : {
        'aktivdir' : True, 
        'endirim' : 0.15
        },
}

m = 'xosh getmisen'

if m in camp:
    if m in disc:
        if disc[m]['aktivdir']:
            print(f'{disc[m]["endirim"] * 100} faiz endirim tetbiq ele')
        else:
            print('endirim aktiv deyil')
    else:
        print(f'{disc["diger"]["endirim"] * 100} faiz endirim tetbiq ele')
else:
    print('bele bir kampaniya movcud deyil')
```
