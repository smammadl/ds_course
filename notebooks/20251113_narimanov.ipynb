# List comprehension

```python
reqemler = [1,3,7,11]
kvadratlar = []

for reqem in reqemler:
    kvadrat = reqem**2
    kvadratlar.append(kvadrat)
    
kvadratlar
```

```python
reqemler = [1,3,7,11]
kvadradlar = [reqem**2 for reqem in reqemler]
kvadradlar
```

```python
reqemler = [-2,-1,1,3,4,7,11]
kvadradlar = [reqem**2 for reqem in reqemler]
kvadradlar
```
