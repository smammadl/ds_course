```python
class Adam:
	species = "Homo sapiens"

	def __init__(self, a_adi, a_soyadi=None, a_yashi=0, a_yashayir=True):
		self.ad = a_adi
		self.soyad = a_soyadi
		self.yash = a_yashi
		self.yashayir = a_yashayir

	def yashlandir(self, nece_yash=1):
		if self.yashayir:
			self.yash += nece_yash
		else:
			print('adam olub')

	def ol(self):
		if not self.yashayir:
			print('qaqash adam olub')
		else:
			self.yashayir = False
			print('adam oldu')

class Talaba(Adam):
	oxuyur = True

t1 = Talaba('Ali', 'Hesenov', 20)
a2 = Adam('Lale')
getattr(a2, 'oxuyur', 'oxumur')
```
