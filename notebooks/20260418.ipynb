```python
import math

class Daire:
	def __init__(self, d_radiusu):
		self.radius = d_radiusu

	def sahe_tap(self):
		self.sahe = math.pi * self.radius**2
		return self.sahe
	
	def perimetr_tap(self):
		self.perimetr = 2 * math.pi * self.radius
		return self.perimetr

d1 = Daire(d_radiusu = 5)
d1.perimetr_tap()
```

```python
from abc import ABC, abstractmethod

class Fiqur(ABC):
	@abstractmethod
	def sahe_tap(self):
		pass
	
	@abstractmethod
	def perimetr_tap(self):
		pass

class Daire(Fiqur):
	def __init__(self, d_radiusu):
		self.radius = d_radiusu

	def sahe_tap(self):
		self.sahe = math.pi * self.radius**2
		return self.sahe
	
	def perimetr_tap(self):
		self.perimetr = 2 * math.pi * self.radius
		return self.perimetr
	
	def equals(self, other):
		if self.radius == other.radius:
			return True
		else:
			return False

	def __eq__(self, other):
		if self.radius == other.radius:
			return True
		else:
			return False
		
	def __repr__(self):
		return f"{self.__class__.__name__}: {self.radius}"

d1 = Daire(d_radiusu = 5)
d2 = Daire(d_radiusu = 5)

# d1.equals(d2)
d1
```
