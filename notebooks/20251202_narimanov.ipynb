```python
arr1 = np.array([
    [1,2,3],
    [4,5,6]
    ])
arr2 = np.array([100,101,102])
arr1 + arr2
```

```python
1. 
[
    [1,2,3],
    [4,5,6]
]
[100,101,102]
# arr1 == (2,3)
# arr2 == (3,)

2. 
[
    [1,2,3],
    [4,5,6]
]
[[100,101,102]]
# arr1 == (2,3)
# arr2 == (1,3)

3. 
[
    [1,2,3],
    [4,5,6]
]
[
    [100,101,102],
    [100,101,102],
]
# arr1 == (2,3)
# arr2 == (2,3)

arr1 + arr2
```

```python
arr1 = np.array([
    [1,2,3],
    [4,5,6]
])
arr2 = np.array([
    [100],
	[101],
])
arr1 + arr2
```

```python
1. 
[
    [1,2,3],
    [4,5,6]
]
[
    [100],
	[101],
]
# arr1 == (2,3)
# arr2 == (2,1)

2. 
[
    [1,2,3],
    [4,5,6]
]
[
    [100, 100, 100],
	[101, 101, 101],
]
# arr1 == (2,3)
# arr2 == (2,3)
```

```python
arr3 = arr1 + arr2
arr3
```
