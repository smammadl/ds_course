```python
import pandas as pd
import pathlib

csv_path = pathlib.Path(r'C:\Users\user\Desktop\div\data\state-abbrevs.csv')
abbrevs = pd.read_csv(csv_path)
pop = pd.read_csv(csv_path.parent/'state-population.csv')
areas = pd.read_csv(csv_path.parent/'state-areas.csv')
```

```python
df_merge_areas = pd.merge(areas, abbrevs, on='state', how='outer')
print(df_merge_areas.shape)
df_merge_areas.head(3)
```

```python
mask_year = pop['year'] == 2013
mask_ages = pop['ages'] == 'total'
pop_yeni = pop[mask_year & mask_ages].reset_index(drop=True)
pop_yeni
```

```python
df_merge_areas.head(3)
```

```python
pop_yeni.head(3)
```

```python
pd.merge(
    df_merge_areas, pop_yeni, 
    left_on='abbreviation', 
    right_on='state/region', 
    how='right').drop(['state/region', 'ages', 'year'], axis='columns')
```

```python
df_merge_pop = pd.merge(pop_yeni, 
                        abbrevs, 
                        left_on='state/region', 
                        right_on='abbreviation',
                        how='left').drop('state/region', axis='columns')
df_merge_pop
```

```python
pd.merge(
    df_merge_pop, 
    df_merge_areas, 
    on='abbreviation', 
    how='left'
    )
```
